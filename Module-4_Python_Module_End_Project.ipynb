```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "C:\\Users\\User\\Downloads\\myexcel.xlsx"
df = pd.read_excel(file_path, sheet_name="myexcel.csv")

# Preprocessing: Fix 'Height' column
np.random.seed(42)
df["Height"] = np.random.randint(150, 181, size=len(df))

# Calculate employee distribution across teams
team_distribution = df["Team"].value_counts()

# Calculate percentage of employees in each team
team_percentage = (team_distribution / len(df)) * 100

# Create a DataFrame to display the results
df_team = pd.DataFrame({"Employee Count": team_distribution, "Percentage": team_percentage})

# Display the team distribution DataFrame
print(df_team)


# Analysis 2: Segregation by position
position_distribution = df["Position"].value_counts()
df_position = pd.DataFrame({"Employee Count": position_distribution})

print(df_position)


# Analysis 3: Predominant age group
age_bins = [20, 25, 30, 35, 40]
age_labels = ["20-25", "26-30", "31-35", "36-40"]
df["Age Group"] = pd.cut(df["Age"], bins=age_bins, labels=age_labels, right=True)
age_group_distribution = df["Age Group"].value_counts()
df_age = pd.DataFrame({"Employee Count": age_group_distribution})

print(df_age)


# Analysis 4: Highest salary expenditure
team_salary = df.groupby("Team")["Salary"].sum().sort_values(ascending=False)
position_salary = df.groupby("Position")["Salary"].sum().sort_values(ascending=False)
print(team_salary)
print(position_salary)

# Analysis 5: Correlation between age and salary
correlation = df["Age"].corr(df["Salary"])
print(f"Correlation between Age and Salary: {correlation:.2f}")







```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "C:\\Users\\User\\Downloads\\myexcel.xlsx"
df = pd.read_excel(file_path, sheet_name="myexcel.csv")

# Preprocessing: Fix 'Height' column
np.random.seed(42)
df["Height"] = np.random.randint(150, 181, size=len(df))

plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Age"], y=df["Salary"], alpha=0.6)
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Correlation between Age and Salary")
plt.show()

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Team distribution
sns.barplot(x=team_distribution.index[:10], y=team_distribution.values[:10], ax=axes[0, 0])
axes[0, 0].set_title("Top 10 Teams by Employee Count")
axes[0, 0].tick_params(axis='x', rotation=90)

# Position distribution
sns.barplot(x=position_distribution.index, y=position_distribution.values, ax=axes[0, 1])
axes[0, 1].set_title("Employees by Position")

# Age group distribution
sns.barplot(x=age_group_distribution.index, y=age_group_distribution.values, ax=axes[1, 0])
axes[1, 0].set_title("Employees by Age Group")

# Salary by position
sns.barplot(x=position_salary.index, y=position_salary.values, ax=axes[1, 1])
axes[1, 1].set_title("Total Salary by Position")
axes[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# Insights summary
insights = """
1. The Cleveland Cavaliers have the highest salary expenditure.
2. The most common position is Shooting Guard (SG).
3. The predominant age group is 20-25 years old.
4. Centers (C) have the highest overall salary allocation.
5. There is no strong correlation between age and salary.
"""
print(insights)
```
